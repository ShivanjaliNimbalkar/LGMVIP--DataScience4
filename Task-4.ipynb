```python
import cv2
import matplotlib.pyplot as plt
```

```python
# import image
img=cv2.imread("apple-1000x1000.jpg")
```

```python
# cv2.imshow("Apple", img)           # it show image
cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# Withoud using cv show image
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
#Convert RGB to Grey Image
#Using cvtColor function of OpenCV.

grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#cv2.destroyAllWindows()
```

```python
#inverted_image = 255 - gray_image
invert_img=cv2.bitwise_not(grey_img)
cv2.imshow("Inverted", invert_img)
cv2.waitKey()
```

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
#cv2.imwrite(sketch.png, sketch_img)
```

```python
inverted_blurred = 255 - blur_img
pencil_sketch = cv2.divide(grey_img, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
#cv2.imshow(sketch image,sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

#Original Image vs Sketch
#We can display the original image and sketch side by side for comparison.

plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
